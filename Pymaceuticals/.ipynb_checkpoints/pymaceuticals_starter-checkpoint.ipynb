```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
clinicaltrial_data = pd.read_csv("data/clinicaltrial_data.csv")
mouse_drug_data = pd.read_csv("data/mouse_drug_data.csv")
mouse_drug_data.head()

# Combine the data into a single dataset
merged_data = pd.merge(clinicaltrial_data, mouse_drug_data, on="Mouse ID")


# Display the data table for preview
merged_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_volume_mean = merged_data.groupby(['Drug','Timepoint']).mean()['Tumor Volume (mm3)']
# Convert to DataFrame
df = pd.DataFrame(tumor_volume_mean)
# Preview DataFrame
df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_volume_sem = merged_data.groupby(['Drug','Timepoint']).sem()['Tumor Volume (mm3)']
# Convert to DataFrame
df_standard_errors = pd.DataFrame(tumor_volume_sem)
# Preview DataFrame
df_standard_errors.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
Merged_data_tumor = merged_data.pivot_table(values='Tumor Volume (mm3)', index = 'Timepoint', columns='Drug')
# Preview that Reformatting worked
Merged_data_tumor.head()
```

```python
# Generate the Plot (with Error Bars)
x_axis = [0,5,10,15,20,25,30,35,40,45]
x_limit = 45
plt.figure(figsize=(8,5))

error = df_standard_errors["Tumor Volume (mm3)"]["Capomulin"]
Capomulin = plt.errorbar(x_axis, Merged_data_tumor["Capomulin"], yerr=error, fmt="o", ls="-",color ="red")
error = df_standard_errors["Tumor Volume (mm3)"]["Infubinol"]
Infubinol = plt.errorbar(x_axis, Merged_data_tumor["Infubinol"], yerr=error, fmt="*", ls="-",color ="green")
error = df_standard_errors["Tumor Volume (mm3)"]["Ketapril"]
Ketapril = plt.errorbar(x_axis, Merged_data_tumor["Ketapril"], yerr=error, fmt="h", ls="-",color ="blue")
error = df_standard_errors["Tumor Volume (mm3)"]["Placebo"]
Placebo = plt.errorbar(x_axis, Merged_data_tumor["Placebo"], yerr=error, fmt="^", ls="-",color ="purple")

plt.title("TumorResponse to Treatment")
plt.xlabel("Time(Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.xlim(0, 45)
plt.ylim(30, 80)

plt.grid(linestyle="-")
plt.legend((Capomulin,Infubinol,Ketapril,Placebo), ("Capomulin", "Infubinol", "Ketapril", "Placebo"))

# Save the Figure
plt.savefig("TumorResponse_to_Treatment.png")
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
metastatic_sites_mean = merged_data.groupby(['Drug','Timepoint']).mean()['Metastatic Sites']
# Convert to DataFrame
df_metastatic_sites = pd.DataFrame(metastatic_sites_mean)
# Preview DataFrame
df_metastatic_sites.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
metastatic_sites_sem = merged_data.groupby(['Drug','Timepoint']).sem()['Metastatic Sites']
# Convert to DataFrame
df_standard_errors_metastatic_sites = pd.DataFrame(metastatic_sites_sem)
# Preview DataFrame
df_standard_errors_metastatic_sites.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
Merged_data_metastatic_sites = merged_data.pivot_table(values='Metastatic Sites', index = 'Timepoint', columns='Drug')
# Preview that Reformatting worked
Merged_data_metastatic_sites.head()
```

```python
# Generate the Plot (with Error Bars)

plt.figure(figsize=(8,5))

error_ms = df_standard_errors_metastatic_sites["Metastatic Sites"]["Capomulin"]
Capomulin_ms = plt.errorbar(x_axis, Merged_data_metastatic_sites["Capomulin"], yerr=error_ms, fmt="o", ls="-", color="red")
error_ms = df_standard_errors_metastatic_sites["Metastatic Sites"]["Infubinol"]
Infubinol_ms = plt.errorbar(x_axis, Merged_data_metastatic_sites["Infubinol"], yerr=error_ms, fmt="h", ls="-", color="green")
error_ms = df_standard_errors_metastatic_sites["Metastatic Sites"]["Ketapril"]
Ketapril_ms = plt.errorbar(x_axis, Merged_data_metastatic_sites["Ketapril"], yerr=error_ms, fmt="*", ls="-", color ="blue")
error_ms = df_standard_errors_metastatic_sites["Metastatic Sites"]["Placebo"]
Placebo_ms = plt.errorbar(x_axis, Merged_data_metastatic_sites["Placebo"], yerr=error_ms, fmt="^", ls="-", color ="purple")

plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met. Sites")
plt.xlim(0, 45)
plt.ylim(0, 4)

plt.grid()
plt.legend((Capomulin_ms,Infubinol_ms,Ketapril_ms,Placebo_ms), ("Capomulin", "Infubinol", "Ketapril", "Placebo"))

# Save the Figure
plt.savefig("Metastatic_Spread_During_Treatment.png")

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_id_mean = merged_data.groupby(['Drug','Timepoint']).count()['Mouse ID']
# Convert to DataFrame
df_mouse_id = pd.DataFrame(mouse_id_mean)
# Preview DataFrame
df_mouse_id.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
Merged_mouse_id_sites = merged_data.pivot_table(values='Mouse ID', index = 'Timepoint', columns='Drug', aggfunc='count')
# Preview the Data Frame
Merged_mouse_id_sites
```

```python
# Generate the Plot (Accounting for percentages)

plt.figure(figsize=(8,5))

#Merged_mouse_id_sites.iloc[0 , 1]
Capomulin_mi = plt.errorbar(x_axis, (Merged_mouse_id_sites["Capomulin"]/Merged_mouse_id_sites.iloc[0 , 1]*100), fmt="o", ls="-", color = "red")
Infubinol_mi = plt.errorbar(x_axis, (Merged_mouse_id_sites["Infubinol"]/Merged_mouse_id_sites.iloc[0 , 1]*100), fmt="h", ls="-", color = "blue")
Ketapril_mi = plt.errorbar(x_axis, (Merged_mouse_id_sites["Ketapril"]/Merged_mouse_id_sites.iloc[0 , 1]*100),  fmt="*", ls="-", color ="green")
Placebo_mi = plt.errorbar(x_axis, (Merged_mouse_id_sites["Placebo"]/Merged_mouse_id_sites.iloc[0 , 1]*100),  fmt="^", ls="-", color ="purple")
plt.title("Survival During Treatment")
plt.xlabel("Time(Days)")
plt.ylabel("Survival Rate(%)")
plt.ylim(30, 110)
plt.xlim(0, 50)


plt.grid()
plt.legend((Capomulin_mi,Infubinol_mi,Ketapril_mi,Placebo_mi), ("Capomulin", "Infubinol", "Ketapril", "Placebo"))

# Save the Figure
plt.savefig("Survival_During_Treatment.png")

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
Percent_change = ((Merged_data_tumor.iloc[-1] - Merged_data_tumor.iloc[0])/Merged_data_tumor.iloc[0])*100

# Display the data to confirm
Percent_change
```

```python
# Store all Relevant Percent Changes into a Tuple
fig = plt.figure(figsize=(8,5))

# Splice the data between passing and failing drugs
x_lable = Percent_change.index
x_values = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]
y_values = [round(Percent_change.loc["Capomulin"]),round(Percent_change.loc["Infubinol"]),round(Percent_change.loc["Ketapril"]),round(Percent_change.loc["Placebo"])]
x_4 = np.arange(4)
# Orient widths. Add labels, tick marks, etc. 

plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")
plt.ylim(-20, 60)
#plt.xlim(0, 50)
plt.bar(x_4, y_values, color=['red' if Percent_change[i] > 0 else 'green' for i in np.arange(len(x_lable))], alpha=0.5, align="center", tick_label=x_values)
plt.grid(linestyle="-")
# Use functions to label the percentages of changes
def percent_label(percent_result):
    count = 0
    for i2 in percent_result:
        if i2 < 0:
            y = -3.5
        else:
            y = 2
        plt.text(count, y, str(round(i2, 1)) + '%', ha='center', color='white')
        #print(i2)
        count +=1
# Call functions to implement the function calls
percent_label(y_values)

# Save the Figure
plt.savefig("Tumor_Change_Over_45_Day_Treatment.png")

# Show the Figure
fig.show()
```
